```python
def find_largest(num_list: list[float]) -> float:
    # """
    # Function to find the largest number in the list.

    # args:
    #     num_list(num_list:[int]): A list containing numeric values (int or float).

    # returns:
    #     float: the largest number in the list

    # raises:
    #     TypeError: If input is not a lit or contains numeric values (int or float)
    #     ValueError: If the list is empty
    # """
    if not isinstance(num_list,list):
        raise TypeError("Input must be a list of numbers float or int.")
    if not num_list:
        raise ValueError("List can't be empty")
    for num in num_list:
        if not isinstance(num,(int,float)):
            raise TypeError("All elements must be numeric.")
        largest = num_list[0]
        for num in num_list:
            if num >largest:
                largest = num
        return largest
#main execution
if __name__ == "__main__":
    try:
        user_input = input("Enter numbers in the list seperate by space or float only")
        num_list = [float(item) for item in user_input.split()]
        result = find_largest(num_list)
        print("Largest numbers in the list is: ", result)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
```

```python
def is_prime_number(number : int) -> bool:
    if not isinstance(number,int):
        raise TypeError("input must be an integer")
    if number < 2:
        raise ValueError("Prime numbers are => 2")
    for i in range (2, int(number**.5) +1):
        if number % i == 0:
            return False
            break
    else:
        return True

if __name__ == "__main__":
    try:
        user_input = input("Enter your number: ")
        if not user_input.strip().isdigit():
            raise TypeError("your input must be an integer")
        number = int(user_input)
        if is_prime_number(number):
            print(f"{number} is a prime number")
        else:
            print(f"{number} is not a prime number")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print("something went wrong" , e)








```

```python
def My_factorial(num : int) -> int:
    if isinstance(num,bool):
        raise TypeError("boolean values are not allowed")
    if not isinstance(num,int):
        raise TypeError("input must be an integer")
    if num < 0:
        raise ValueError("factorial is not defined for negetive numbers")
    factorial = 1
    for i in range(1,num+1):
        factorial = factorial * i
    return factorial
if __name__ == "__main__":
    try:
        number = input("enter your number to find factorial: ")
        number = int(number)
        print("factorial",My_factorial(number))
    except TypeError as e:
        print(e)
    except ValueError as e:
        print("input must be an int")
```

```python
def my_palindrome(text: str) -> bool:
    if not isinstance(text,str):
        raise("Input must be a string")
    if text.strip() == "":
        raise TypeError ("Input text can not be empty")
    else:
        text = text.strip().lower()
    reversed_string = ""
    for char in text:
        reversed_string = char + reversed_string
    return text == reversed_string

if __name__ == "__main__":
    try:
        user_input = input("Enter your text to see if its a palin.... or not").strip().lower()
        if my_palindrome(user_input):
            print(f"{user_input} is a palindrome")
        else:
            print(f"{user_input} is not a palindrome")
    except(TypeError,ValueError) as e:
        print("Error : ",e)
```

```python
def second_largest(lst):
    if len(lst) < 2:
        return None
    largest = float('-inf')
    second = float('-inf')
    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num
    if second == float('-inf'):
        return None
    return second
user_input = input("Enter numbers separated by space: ")
numbers = []
for x in user_input.split():
    numbers.append(int(x))
result = second_largest(numbers)
if result is not None:
    print("Second largest number is:", result)
else:
    print("No second largest number found")
```
